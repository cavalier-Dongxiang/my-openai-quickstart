````python
from langchain_core.messages import BaseMessage, HumanMessage, ToolMessage
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langgraph.graph import END, StateGraph, START


#  LLM 
def create_agent(llm, tools, tool_message: str, custom_notice: str=""):
    """"""
    # 
    prompt = ChatPromptTemplate.from_messages(
        [
            (
                "system",
                "You are a helpful AI assistant, collaborating with other assistants."
                " Use the provided tools to progress towards answering the question."
                " If you are unable to fully answer, that's OK, another assistant with different tools "
                " will help where you left off. Execute what you can to make progress."
                " If you or any of the other assistants have the final answer or deliverable,"
                " prefix your response with FINAL ANSWER so the team knows to stop."
                "\n{custom_notice}\n"
                " You have access to the following tools: {tool_names}.\n{tool_message}\n\n",
            ),
            MessagesPlaceholder(variable_name="messages"),  # 

        ]
    )

    # 
    prompt = prompt.partial(tool_message=tool_message, custom_notice=custom_notice)
    prompt = prompt.partial(tool_names=", ".join([tool.name for tool in tools]))
      
    # 
    return prompt | llm.bind_tools(tools)
````

from typing import Literal
from langgraph.graph import StateGraph, MessagesState

def should_continue(state: MessagesState) -> Literal["tools", "__end__"]:
    messages = state["messages"]
    last_message = messages[-1]
    if last_message.tool_calls:
        return "tools"
    return "__end__"

def call_model(state: MessagesState):
    messages = state["messages"]
    response = model_with_tools.invoke(messages)
    return {"messages": [response]}

# 
workflow = StateGraph(MessagesState)

# 
workflow.add_node("agent", call_model)
workflow.add_node("tools", tool_node)

workflow.add_edge("__start__", "agent")  #  agent 
workflow.add_conditional_edges("agent", should_continue)  # 
workflow.add_edge("tools", "agent")  #  agent 

app = workflow.compile()  #

````python
#  workflow AgentState 
workflow = StateGraph(AgentState)

# 
workflow.add_node("Researcher", research_node)
workflow.add_node("Chart_Generator", chart_node)
workflow.add_node("call_tool", tool_node)
````


#### 



#### 
- `router` 
- `tool_calls` `"call_tool"`
-  `"FINAL ANSWER"` `"__end__"` 
-  `"continue"`

````python
from typing import Literal

# 
def router(state) -> Literal["call_tool", "__end__", "continue"]:
    messages = state["messages"]  # 
    last_message = messages[-1]  # 
    
    #  "call_tool"
    if last_message.tool_calls:
        return "call_tool"
    
    #  "FINAL ANSWER" "__end__" 
    if "FINAL ANSWER" in last_message.content:
        return "__end__"
    
    #  "continue"
    return "continue"
````

#### 

````python
#  "Researcher"  router 
workflow.add_conditional_edges(
    "Researcher",
    router,  # 
    {
        "continue": "Chart_Generator",  #  router  "continue" Chart_Generator
        "call_tool": "call_tool",  #  router  "call_tool"
        "__end__": END  #  router  "__end__"
    },
)

#  "Chart_Generator" 
workflow.add_conditional_edges(
    "Chart_Generator",
    router,  #  router 
    {
        "continue": "Researcher",  #  router  "continue" Researcher
        "call_tool": "call_tool",  #  router  "call_tool"
        "__end__": END  #  router  "__end__"
    },
)

#  "call_tool" sender
#  sender 
workflow.add_conditional_edges(
    "call_tool",
    lambda x: x["sender"],  #  sender 
    {
        "Researcher": "Researcher",  #  sender  Researcher Researcher
        "Chart_Generator": "Chart_Generator",  #  sender  Chart_Generator Chart_Generator
    },
)
````

````python
#  START  Researcher 
workflow.add_edge(START, "Researcher")

# 
graph = workflow.compile()
````
